# Create Datasets (Linearly and Non-linearly separable)

```python
data_a_anchor = [4, 4]
data_b_anchor = [-4, 4]
data_size = 6
mu, sigma = 0, 2.3
noise = np.random.normal(mu, sigma, 1000)
data_a = []
data_b = []
data_c = []
data_d = []
for ii in xrange(data_size):
    a_point = [data_a_anchor[0] + np.random.choice(noise), data_a_anchor[1] + np.random.choice(noise)]
    b_point = [data_b_anchor[0] + np.random.choice(noise), data_b_anchor[1] + np.random.choice(noise)]
    data_a.append(a_point)
    data_b.append(b_point)
data_a = np.array(data_a)
data_b = np.array(data_b)
data_c = np.concatenate((data_a, -1.0 * data_a))
data_d = np.concatenate((data_b, -1.0 * data_b))
example_data_a = np.asarray([[1, 6], [7, 2], [8, 9], [9, 9]])
example_data_b = np.asarray([[2, 1], [2, 2], [2, 4], [7, 1]])
plt.scatter(data_a[:,0], data_a[:,1])
plt.scatter(data_b[:,0], data_b[:,1])
plt.title('Linearly Separable Data')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('figures/linear_data.png')
plt.show()

plt.scatter(data_c[:,0], data_c[:,1])
plt.scatter(data_d[:,0], data_d[:,1])
plt.title('Non-linearly Separable Data')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('figures/non_linear_data.png')
plt.show()

plt.scatter(example_data_a[:,0], example_data_a[:,1])
plt.scatter(example_data_b[:,0], example_data_b[:,1])
plt.title('Example Data')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('figures/example_data.png')
plt.show()
```

# Perceptron and LMS Models

```python
def train_classifier(data, alpha, epochs):
    w_vec = [0.1, 0.1, 0.1]
    err_vecs = [] # log all the errors
    
    epoch_classifications = np.zeros(epochs) # count incorrect classificaitons
    epoch_errors = np.zeros(epochs) # add error magnitudes
    for ii in xrange(epochs):
        def classify(point):
            return np.dot(w_vec, point)

        for point in data:
            # Incorrectly classified point
            classification = classify(point)
            if classification < 0:
                w_vec = np.add(w_vec, np.multiply(alpha, point))
                epoch_classifications[ii] += 1
                epoch_errors[ii] += classification
    
    return w_vec, epoch_classifications, epoch_errors


def train_lms(data):
    yT = data.T
    yTy = np.dot(yT, data)
    yTy_inv = np.linalg.inv(yTy)
    b = np.transpose(np.ones(len(data)))
    yTb= np.dot(yT, b)
    vec = np.dot(yTy_inv, yTb)
    
    return vec


def normalize_data(a, b):
    # "normalize" data
    a = np.insert(a, 0, 1, axis=1)
    b = np.insert(b, 0, 1, axis=1)
    data = np.concatenate((a, -1.0 * b))
    
    return data


def vector_norm(x):
        length = np.linalg.norm(x)
        if length != 0:
            return x / length
        else:
            return [0, 0]
        
        
def get_boundary_intercepts(weights, scale=1):
    """Decision boundary plotting code from: 
    https://stats.stackexchange.com/questions/169343/how-to-plot-the-or-function-along-with-the-decision-boundary-of-a-perceptron
    This link was sent by Rishabh"""
    # w_1*x + w_2*y + b = 0
    b, w1, w2 = [float(w) for w in weights]
    
    # x and y intercepts
    x = -b / w1
    y = -b / w2

    return x, y


def plot_line(p1, p2, color, label):
    ax = plt.gca()
    xmin, xmax = ax.get_xbound()

    if(p2[0] == p1[0]):
        xmin = xmax = p1[0]
        ymin, ymax = ax.get_ybound()
    else:
        ymax = p1[1]+(p2[1]-p1[1])/(p2[0]-p1[0])*(xmax-p1[0])
        ymin = p1[1]+(p2[1]-p1[1])/(p2[0]-p1[0])*(xmin-p1[0])

    l = mlines.Line2D([xmin,xmax], [ymin,ymax], color=color, label=label)
    ax.add_line(l)
    return l
    

def plot_figures(a, b, perceptron_weights, epoch_classifications, epoch_errors, lms_weights, tag=None):
    plt.scatter(a[:,0], a[:,1], color='red')
    plt.scatter(b[:,0], b[:,1], color='blue')
    
    print 'Perceptron:', perceptron_weights
    print 'LMS:', lms_weights
    
    # plot decision boundaries
    perceptron_x, perceptron_y = get_boundary_intercepts(perceptron_weights)#, scale=max([x_max, y_max]))
    lms_x, lms_y = get_boundary_intercepts(lms_weights)#, scale=max([x_max, y_max]))

    plot_line([perceptron_x, 0], [0, perceptron_y], color='green', label='Perceptron')
    plot_line([lms_x, 0], [0, lms_y], color='orange', label='LMS')
    
    plt.title('Perceptron vs LMS Classification Decision Boundaries')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    filename = 'decision_boundaries.png'
    if tag is not None:
        filename = 'decision_boundaries_{}.png'.format(tag)
    plt.savefig('figures/' + filename)
    plt.show()
    

    # plot errors
    plt.plot(np.absolute(epoch_errors))
    plt.title('Total Error Magnitude vs Iteration')
    plt.ylabel('|sum(a*Y)|')
    plt.xlabel('iteration')
    filename = 'epoch_errors.png'
    if tag is not None:
        filename = 'epoch_errors_{}.png'.format(tag)
    plt.savefig('figures/' + filename)
    plt.show()
    
    plt.plot(np.absolute(epoch_classifications))
    plt.title('Incorrect Classifications vs Iteration')
    plt.ylabel('incorrect classifications')
    plt.xlabel('iteration')
    plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True)) # force integer ticks on y axis
    filename = 'epoch_classifications.png'
    if tag is not None:
        filename = 'epoch_classifications_{}.png'.format(tag)
    plt.savefig('figures/' + filename)
    plt.show()
```

# Self Organizing Maps

```python
class SOM(object):
    
    def __init__(self):
        self.lattice = None
    
    
    def train(self, data, lattice_length, iterations, time_steps,\
              plot_iterations=[2, 5, 10, 50], initial_learning_rate=0.5, initial_neighborhood_size=1.0,\
             data_name=None, save_figs=False):
        self.time_steps = time_steps
        self.initial_learning_rate = initial_learning_rate
        self.initial_neighborhood_size = initial_neighborhood_size
        # Logs
        self.h_log = []
        self.lr_log = []
        self.n_log = []
        
        # square lattice of neurons, initialized with random weights uniformly from [0, 1)
        # the neurons are just weights
        self.lattice = np.random.rand(lattice_length, lattice_length, 2) # TODO: can abstract data dimmensionality by replacing 2 with parameter
        if data_name is not None:
            title = '{}, Iteration {}'.format(data_name, 0)
        else:
            title = 'Iteration {}'.format(0)
        self.plot_lattice(title, save_figs)
        
        
        t = 0
        for iteration in xrange(iterations):
            lr = self.learning_rate(t) # learning rate, decays with time
            n = self.neighborhood_size(t) # neighborhood size, decays with time
            for point in data:
                # Find the unit w_i that best matches the point
                w_min = None
                dist = None
                for ii in xrange(lattice_length):
                    for jj in xrange(lattice_length):
                        w_i = self.lattice[ii][jj]
                        l2 = np.linalg.norm(w_i - point) # l2 norm a.k.a euclidian distance
                        if dist is None or l2 < dist:
                            w_min = [ii, jj]
                            dist = l2
                            
                # Update each neuron, effect of update decays as distance from best neuron increases
                for ii in xrange(lattice_length):
                    for jj in xrange(lattice_length):
                        dist = self.distance(w_min, [ii, jj]) # lateral distance between the two neurons (in lattice)
                        h = self.kernel(t, n, dist) # kernel, function of neighborhood size decay
                        w_k = self.lattice[ii][jj]
                        self.lattice[ii][jj] = w_k + lr * h * (point - w_k)
                        
                        self.h_log.append(np.asarray([dist, h, jj, ii, t]))
            
            #self.lattice = neuron_lattice # save each iteration, so we can plot it at any point
                        
            t += 1
            self.lr_log.append(lr)
            self.n_log.append(n)
            
            # Plot lattics
            if t in plot_iterations:
                title = None
                if data_name is not None:
                    title = '{}, Iteration {}'.format(data_name, t)
                else:
                    title = 'Iteration {}'.format(t)
                self.plot_lattice(title, save_figs)
    
    
    def learning_rate(self, t):
        return self.initial_learning_rate * math.exp(-t*1.0/self.time_steps)
    
    
    def kernel(self, t, neighborhood_size, distance):
        return math.exp((-(distance**2) * 1.0)/
                        (2.0 * (neighborhood_size**2)))
    
    def distance(self, p1, p2):
        """p1 and p2 are two cell list coordinates e.g. [1, 3]"""
        assert len(p1) == 2 and len(p2) == 2
        return math.fabs(p1[0] - p2[0]) + math.fabs(p1[1] - p2[1])
    
    
    def neighborhood_size(self, t):
        return self.initial_neighborhood_size * math.exp(-t*1.0/self.time_steps)
    
    
    def plot_lattice(self, title=None, save_figs=False):
        """Assumes square net."""
        assert self.lattice is not None
        
        for ii in xrange(len(self.lattice)):
            for jj in xrange(len(self.lattice[0])):
                def connectpoints(p1,p2):
                    plt.plot([p1[0],p2[0]],[p1[1],p2[1]],'k-')

                p1 = self.lattice[ii][jj]
                try:
                    p2 = self.lattice[ii+1][jj]
                    connectpoints(p1, p2)
                except IndexError:
                    pass
                try:
                    p3 = self.lattice[ii][jj+1]
                    connectpoints(p1, p3)
                except IndexError:
                    pass

        plt.xlim((-4, 4))
        plt.ylim((-4, 4))
        if title is not None:
            plt.title(title)
            if save_figs:
                plt.savefig('figures/{}.png'.format(title))
        plt.show()
        
        
    def plot_logs(self, save_figs):
        plt.plot(self.lr_log)
        plt.title('Learning Rate vs Iteration')
        plt.ylabel('n')
        plt.xlabel('iteration')
        if save_figs:
            plt.savefig('figures/lr_vs_time.png')
        plt.show()
        
        plt.plot(self.n_log)
        plt.title('Neighborhood Size vs Iteration')
        plt.ylabel('sigma')
        plt.xlabel('iteration')
        if save_figs:
            plt.savefig('figures/sigma_vs_time.png')
        plt.show()
        
#         assert len(self.h_log[0]) == 5
#         h_log = np.asarray(self.h_log)
#         h_log = h_log[(h_log[:,4] == 20) & (h_log[:,2] == 3) & (h_log[:,3] == 3)]
#         print h_log
#         plt.plot(h_log[:,0], h_log[:,1])
#         plt.title('H vs Time')
#         plt.show()
```
